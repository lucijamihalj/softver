```python
import numpy as np
broj_bacanja = 1000
broj_pokusa= 100000
```

```python
pokus = np.random.uniform(size=(broj_bacanja, broj_pokusa))
glave = pokus < 0.55
```

```python
import matplotlib.pyplot as plt
p = np.arange(0,1,0.01)
def P(x):
    glave = pokus < x
    broj_glava = glave.sum(axis=0)
    broj_pisama = broj_bacanja - broj_glava
    vjerojatnost = (broj_glava > broj_pisama).mean()
    return vjerojatnost
%matplotlib inline
fig, ax=plt.subplots()
for i in p:
    ax.scatter(i,P(i))
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
x, y = np.meshgrid(np.linspace(-2, 2), np.linspace(0, 4))
z = np.sin(np.sqrt(x**2 + y**2))
z
```

```python
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_surface(x, y, z, rstride=5, cstride=4, linewidth=10, cmap='terrain', facecolor='r')
```

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(1,1,1,projection='3d')
p = ax2.plot_wireframe(x,y,z,rstride=4, cstride=4)
```

```python
# koordinatne osi = spines
fig, ax = plt.subplots()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.set_xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
ax.set_xticklabels([r'-$\pi$', r'-$\frac{\pi}{2}$', 0, r'$\frac{\pi}{2}$', r'$\pi$'], fontsize=18)
ax.spines['bottom'].set_position(('data',0)) # pozicija x-osi je x=0
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # pozicija y-osi je y=0
xx = np.linspace(-np.pi, np.pi, 1000)
ax.plot(xx, np.sin(xx), label='sin', color='r')
ax.plot(xx, np.cos(xx), label='cos', color='b')
ax.set_ylim(-1,2)
ax.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{2}}{3}$', (2*np.pi/3,np.sin(2*np.pi/3)),(2*np.pi/3-1,1.3), arrowprops=dict(arrowstyle='->'))
ax.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', (2*np.pi/3, -1/2), (2*np.pi/3-1, -1), arrowprops=dict(arrowstyle='->'))
ax.plot([2*np.pi/3,2*np.pi/3], [np.sin(2*np.pi/3), 0],marker='o',color='b',linestyle='--')
ax.plot([2*np.pi/3,2*np.pi/3], [0, -1/2],marker='o',color='r',linestyle='--')
ax.legend()
```
