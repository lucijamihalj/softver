```python
rez=dict()
```

```python
def rjecnik(l,f):
    rez=dict()
    for element in l:
        rez[element]=f(element)
    return rez
```

```python
l=[1, 2, 3]
def f(x):
    return x**2
```

```python
rjecnik(l,f)
```

# Zadatak 5

```python
import sympy as sp
def f(n,x,k):
    y = sp.Symbol('y')
    lista=[]
    Bell = sp.functions.combinatorial.numbers.bell(n, sp.Symbol('y'))
    print(Bell)
    for i in range(k):
        lista.append(sp.Derivative(Bell, sp.Symbol('y'),i).doit().n(30, subs={y:x}))
   
    return lista
    
```

```python
f(10,25,3)
```
